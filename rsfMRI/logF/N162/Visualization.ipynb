```python
from nilearn import plotting
from nilearn import image
import matplotlib.pyplot as plt
```

```python
oldfont = plt.rcParams['font.sans-serif']
plt.rcParams['font.sans-serif'] = ['FreeSans']

tmap = image.load_img("spmT_0001.nii")
tmap = image.math_img("np.where(i<0,0,i)",i=tmap)

plotting.plot_stat_map(tmap, 
                       threshold=5.212,  # FWE < 0.05
                       cut_coords=(0, -21, 10))
plt.savefig("logF.png")
```
