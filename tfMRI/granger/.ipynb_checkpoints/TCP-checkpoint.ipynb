# Load Pandas and Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Now load the Power 2011 parcellation

```python
power = pd.read_csv("../rsfmri/bin/power_2011.csv")
power
```

For each ROI in Power, load the corresponding timeseries for Subject 001

```python
data = []
for roi in list(power['ROI']):
    entry = power.loc[power['ROI'] == roi]
    net = int(entry['Network'])
    net = max(0, net)
    mat = pd.read_csv("../rsfmri/sub-001/func/region_%03d_network_%02d.txt" % (roi, net), delim_whitespace=True)
    vec = np.mean(mat, axis=0)
    row = list(vec)
    data.append(row)
```

Turn data in t X p array

```python
mdata = np.array(data)
mdata
```

 Plot the data for inspection
